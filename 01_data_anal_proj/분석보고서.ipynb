# 항공사 지연 요인 분석 보고서

## 1. 프로젝트 개요
- **분석 목적**: 항공사의 지연 요인(운항사, 관제, 기상)을 파악하여 개선 방향 제시
- **데이터 출처**: 항공 운항 지연률 데이터 (총 33,468건)
- **분석 대상 변수**:  
  - `pct_carrier_delay`: 운항사 요인
  - `pct_atc_delay`: 관제 요인
  - `pct_weather_delay`: 기상 요인
  - `airline`: 항공사명

```python
# 데이터 불러오기
import pandas as pd

df = pd.read_csv("data/airline_stats.csv")  # 파일명은 상황에 맞게 수정

# 기본 정보 확인
df.info()
```

```python
# 데이터 샘플 보기
df.head()
```

## 2. 데이터 전처리
### 2.1 결측치 탐색
- 결측치가 있는 변수: `pct_carrier_delay`, `pct_atc_delay`, `pct_weather_delay` (각 28건)

### 2.2 결측치 처리
- **방안 A**: 행 제거 (`dropna`)
- **방안 B**: 평균으로 대체 (`fillna`)
- **방안 C**: 항공사별 평균으로 대체 (`groupby().transform()`)

*세 가지 방식의 처리 후 분포 비교 및 통계 요약 포함*


```python
# 결측치 개수
missing_count = df.isnull().sum()

# 전체 대비 결측치 비율(%)
missing_ratio = (missing_count / len(df)) * 100

# 결측치 요약표 출력
missing_df = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing Ratio (%)': missing_ratio
}).sort_values(by='Missing Count', ascending=False)

missing_df[missing_df['Missing Count'] > 0]
```

```python
#폰트 지정
from matplotlib import rcParams
rcParams['font.family'] = 'Gulim'
rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt

# 결측치 있는 컬럼만 시각화
missing_df = missing_df[missing_df['Missing Count'] > 0]

plt.figure(figsize=(8, 4))
plt.barh(missing_df.index, missing_df['Missing Count'], color='salmon')
plt.xlabel('결측치 개수')
plt.title('열별 결측치 분포')
plt.tight_layout()
plt.show()
```

```python
# A. 결측값이 있는 행 제거
df_dropna = df.dropna()

# B. 전체 평균으로 결측값 대체
df_filled = df.copy()
for col in ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']:
    df_filled[col] = df_filled[col].fillna(df_filled[col].mean())

# C. 항공사별 평균으로 결측값 대체
df_group_filled = df.copy()
for col in ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']:
    df_group_filled[col] = df_group_filled.groupby('airline')[col]\
                                          .transform(lambda x: x.fillna(x.mean()))

```

```python
# 수치형 변수만 추출
num_cols = ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']

# describe 비교
desc_dropna = df_dropna[num_cols].describe().add_prefix('dropna_')
desc_filled = df_filled[num_cols].describe().add_prefix('filled_')
desc_group = df_group_filled[num_cols].describe().add_prefix('group_')

# 세 개 결과를 하나로 합치기
summary_all = pd.concat([desc_dropna, desc_filled, desc_group], axis=1)
import pandas as pd
from IPython.display import display
display(summary_all)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 3, figsize=(15, 10))
datasets = [('dropna', df_dropna), ('filled', df_filled), ('group', df_group_filled)]

for i, col in enumerate(num_cols):
    for j, (label, data) in enumerate(datasets):
        axes[i, j].hist(data[col], bins=30, color='skyblue', edgecolor='black')
        axes[i, j].set_title(f'{label}: {col}')
        axes[i, j].set_xlabel(col)
        axes[i, j].set_ylabel('Freq')

plt.tight_layout()
plt.show()
```

- 평균 대체는 데이터 수를 유지하면서 평균은 같지만 분산이 약간 줄어든다.
- 그룹 평균 대체는 항공사별 특징을 유지하면서 보다 정교한 대체 방식이다.
- dropna는 샘플 수가 줄어들어 외곽값의 영향이 커질 수 있다.

선택된 처리 방식: 항공사별 그룹 평균 대체

결론: 본 분석에서는 항공사별 그룹 평균을 이용한 결측치 보간 방식을 적용하였다. <br>
이는 데이터 손실을 최소화하고 항공사별 지연률 특성을 보존하여, 이후 분석의 정확도를 높일 수 있다.

```python
# 수치형 데이터 pct_carrier_delay	pct_atc_delay	pct_weather_delay
# 범주형 데이터 airline
df.describe(), df['airline'].unique()
```

## 3. 탐색적 데이터 분석 (EDA)
### 3.1 변수별 분포
- 히스토그램 및 상자그림(boxplot) 시각화
- 이상치(outlier) 여부 확인

### 3.2 항공사별 비교
- 항공사별 평균 지연율 비교 (막대그래프)
- 항공사별 상자그림 비교

```python
# 시각화에 필요한 라이브러리를 불러옵니다.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 분석할 수치형 변수 목록
numerical_cols = ['pct_carrier_delay', 'pct_atc_delay', 'pct_weather_delay']
```

```python
# 각 수치형 변수에 대해 항공사별 상자 그림 그리기
for col in numerical_cols:
    plt.figure(figsize=(14, 8))
    # x축에 범주형 변수, y축에 수치형 변수를 지정하는 것이 핵심입니다.
    sns.boxplot(data=df, x='airline', y=col)
    
    plt.title(f'항공사별 "{col}" 분포 (Box Plot)', fontsize=16)
    plt.xlabel('항공사', fontsize=12)
    plt.ylabel(f'{col} (%)', fontsize=12)
    plt.xticks(rotation=30) # 항공사 이름이 길 경우를 대비해 살짝 회전
    plt.show()
```

```python
# --- 3개 변수의 상자 그림을 한 번에 그려서 비교하기 ---
plt.figure(figsize=(15, 8))
sns.boxplot(data=df[numerical_cols])
plt.title('지연 유형별 분포 및 이상치 확인 (Box Plot)', fontsize=16)
plt.ylabel('지연율 (%)', fontsize=12)
plt.show()


# --- 코드로 이상치 직접 확인하기 ---
# 예시: 'pct_atc_delay' 변수의 이상치 찾아보기
Q1 = df['pct_atc_delay'].quantile(0.25)
Q3 = df['pct_atc_delay'].quantile(0.75)
IQR = Q3 - Q1 # 사분위수 범위 (Interquartile Range)

# 이상치의 경계 계산 (일반적으로 IQR의 1.5배수를 사용)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 경계를 벗어나는 데이터 필터링
outliers = df[(df['pct_atc_delay'] < lower_bound) | (df['pct_atc_delay'] > upper_bound)]

print(f"--- 'pct_atc_delay'의 이상치 정보 ---")
print(f"IQR(사분위수 범위): {IQR:.2f}")
print(f"이상치 상한 경계: {upper_bound:.2f}")
print(f"총 이상치 개수: {len(outliers)}")
print("\n가장 큰 이상치 5개:")
print(outliers.sort_values(by='pct_atc_delay', ascending=False).head())
```
