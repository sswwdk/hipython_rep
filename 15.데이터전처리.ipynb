```python
w_df.columns
```

```python
w_df.head(3)
```

```python
w_df.tail(3)
```

```python
#1) day로 언피봇 melt()
w_melt = w_df.melt(
    id_vars=['id','year','month','element'],
    var_name='day',
    value_name='temp'
)

print(w_melt)
```

```python
#2) element를 피봇팅 > tmax, tmin 컬럼 생성
w_tidy = w_melt.pivot_table(
    index = ['id', 'year','month', 'day'],
    columns = 'element',
    values = 'temp'
)

w_tidy.head()
```

```python
# element 랑 day가 겹쳐서 평탄화
w_tidy_flat = w_tidy.reset_index()

w_tidy_flat.head()
```

```python
#2-2) element를 피봇팅 > tmax, tmin 컬럼 생성, element랑 day가 겹쳐서 평탄화를 한꺼번에.
w_tidy_once = w_melt.pivot_table(
    index = ['id', 'year','month', 'day'],
    columns = 'element',
    values = 'temp'
).reset_index()

w_tidy_once.head()
```

# 2. apply()   
   
함수를 시리즈, 데이터플레임에 적용하기

```python
# 1. 사용자 정의 함수

def my_sq(x):
    return x**2

# 2. 데이터프레임.apply(my_sq)
```

```python
import pandas as pd

df = pd.DataFrame({'a': [10, 20, 30],
                   'b': [20, 30 ,40]})
df
```

```python
df.apply(my_sq)
```

```python
# 데이터 프레임 함수
dir(df)
```

```python
# 벡터화된 연산 
df['a'] ** 2
```

```python
my_sq(df['a']) # 이렇게 해도 되지만~~
```

```python
df.iloc[0].apply(my_sq) # 한 행 apply
```

```python
# 매개변수 2개 함수 정의
def my_exp(x, e):
    return x**e

my_exp(2,3)
```

```python
df['a'].apply(my_exp,e=3)
```

```python
df.iloc[0].apply(my_exp, e=2)
```

```python
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
print(df)
```

```python
def age_category(age):
    if age < 30:
        return '20대'
    elif age < 40:
        return '30대'
    else:
        return'할아버지'

df['Age'].apply(age_category)
```

```python
# Category 칼럼으로 추가
df['Category'] = df['Age'].apply(age_category)
df
```

```python
def poor_man(Salary):
    if Salary < 6000:
        return "i'm Poor"
    else:
        return "i'm Okay"
```

```python
df['State'] = df['Salary'].apply(poor_man)  # 열 추가 방법 1
df
```

```python
def bonus(x):
    return x*0.03

df.insert(loc=3, column='Bonus',value=bonus(df['Salary'])) # 열 추가 방법 2
# loc=3 >>  column 3 뒤에
df
```

```python
import numpy as np
```

```python
# 여러개를 적용해서 합할 때  agg()
# 줄바꿈을 해서 사용하고 싶을 때 \(역슬레시) 하면 된다.
df.groupby('continent')['lifeExp']\
.agg([np.std, np.mean, np.min, np.max])
```

```python
agg_df = df.groupby('continent')['lifeExp']\
    .agg(
        {
            'lifeExp': 'mean',
            'pop': 'median',
            'gdpPercap':'median'
        }
    )

agg_df
```

```python
agg_df = df.groupby('continent')['lifeExp']\
 .agg(
    {
        'lifeExp': 'mean',
        'pop': 'median',
        'gdpPercap':'median'
    }
)

agg_df
```
