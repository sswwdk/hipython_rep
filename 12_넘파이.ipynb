```python
# 1차원 배열 생성
arr = np.array([0, 1, 2, 3, 4, 5])

# 단일 인덱싱
print("첫 번째 요소:", arr[0])      # 0
print("마지막 요소:", arr[-1])      # 5

# 슬라이싱
print("처음부터 3개:", arr[:3])     # [0 1 2]
print("2번째부터 4번째까지:", arr[2:5])  # [2 3 4]
print("처음부터 끝까지 2단계씩:", arr[::2])  # [0 2 4]

# 조건 인덱싱
print("3보다 큰 요소:", arr[arr > 3])  # [4 5]
```
