```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,6))
plt.title('5일간 주식 가격') # 그래프 제목 (중앙 맨 위)
plt.plot(dates, prices, marker='o',label = '주식 가격', color = 'blue')
# x축, y축, marker , legend label, color
plt.xlabel('날짜') # x축 이름
plt.ylabel('가격') # y축 이름
plt.legend() # 왼쪽 위에 레전드 # plt.plot 뒤에 와야 된다. 아니면 이름이 없어짐
plt.grid() # 그래프 그리드 
plt.ylim(95,110) # y축 범위
plt.xlim(0,4) # x축 범위
```

```python
plt.subplot(1, 2, 1)
plt.title('Stock Price - Subplot 1')
plt.plot(dates, prices, marker='o', color='green')
plt.xticks(rotation=45) # x축 요소 40도 돌려주기


plt.subplot(1, 2, 2)
plt.title('Stock Price - Subplot 2')
plt.bar(dates, prices, color='orange')
plt.xticks(rotation=45)
```

# Seaborn 데이터셋

```python
import seaborn as sns
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
plt.hist(data = tips_df, x ='total_bill', bins=20) #bins
```

```python
fig =  plt.figure(figsize=(12,5))
axes1 = fig.add_subplot(1,2,1) # total_bill
axes2 = fig.add_subplot(1,2,2) #tips
axes1.hist(data = tips_df, x = 'total_bill', bins=20)
axes2.hist(data = tips_df, x = 'tip', color = 'orange')
plt.tight_layout
plt.show
```

```python
plt.subplot(1,2,1)
plt.hist(data = tips_df, x ='total_bill', bins=20) #bins


plt.subplot(1,2,2)
plt.hist(data = tips_df, x = 'tip', color = 'yellow')
```

```python
#boxplot
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)  #total_bill
axes.boxplot(data=tips_df, x='total_bill')
plt.show()
```

```python
# total_bill(Male vs Female)
fig = plt.figure(figsize=(12,5)) 
axes = fig.add_subplot(1,1,1) # total_bill
axes.boxplot(
    x =[
        tips_df[tips_df['sex']=='Female']['total_bill'],
        tips_df[tips_df['sex']=='Male']['total_bill']        
    ],
    labels = ['Female', 'Male']
)   
plt.show
```

```python
# tip (Male vs Female)
fig = plt.figure(figsize=(12,5)) 
axes = fig.add_subplot(1,1,1) # total_bill
axes.boxplot(
    x =[
        tips_df[tips_df['sex']=='Female']['tip'],
        tips_df[tips_df['sex']=='Male']['tip']        
    ],
    labels = ['Female', 'Male']
)   
plt.show
```

```python
tips_df.head()
```

```python
tips_df['size'].unique()
```

```python
# all size : tip (Male vs Female)
fig = plt.figure(figsize=(12,5)) 
axes = fig.add_subplot(1,1,1) # tip
axes.boxplot(
    x =[
        tips_df[tips_df['size']==1]['tip'],
        tips_df[tips_df['size']==2]['tip'],
        tips_df[tips_df['size']==3]['tip'], 
        tips_df[tips_df['size']==4]['tip'],
        tips_df[tips_df['size']==5]['tip'],
        tips_df[tips_df['size']==6]['tip'],      
    ],
    labels = ['1','2','3','4','5','6']
)
axes.set_title('인원 별 팁')

plt.show
```

```python
# size 2 : tip (Male vs Female)
fig = plt.figure(figsize=(12,5)) 
axes = fig.add_subplot(1,1,1) # tip
axes.boxplot(
    x=[
        tips_df[(tips_df['size'] == 2) & (tips_df['sex'] == 'Male')]['tip'],
        tips_df[(tips_df['size'] == 2) & (tips_df['sex'] == 'Female')]['tip']
    ],
    labels=['Male', 'Female']
)
axes.set_title('2인: 성별에 따른 팁')

plt.show
```

```python
# scatter, 산점도
fig = plt.figure()
axes = fig.add_subplot(1,1,1)

axes.scatter(
    tips_df['total_bill'], tips_df['tip']
)

axes.set_title('Scatter')
axes.set_xlabel('total_bill')
axes.set_ylabel('tip')

plt.show
```

```python
colors = {"Female": "red", "Male": "blue"}
# 다변량 표현
# 성별에 따라 - 점의 색깔
# 값의 크기에 따라 - 점의 크기
fig = plt.figure()
axes = fig.add_subplot(1,1,1)

axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s= tips_df['size']**2*10, #tablesize
    c= tips_df['sex'].map(colors)
)

axes.set_title('Scatter')
axes.set_xlabel('total_bill')
axes.set_ylabel('tip')

plt.show
```

```python
colors = {"Female": "red", "Male": "blue"}
# 다변량 표현
# 성별에 따라 - 점의 색깔
# 값의 크기에 따라 - 점의 크기
fig = plt.figure()
axes = fig.add_subplot(1,1,1)

axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s= tips_df['tip']**2*10, 
    c= tips_df['sex'].map(colors),
    alpha=0.5
)
plt.show
```

```python
colors = {"Female": "#ff00ad", "Male": "#000080"}
tips_df['tip_pct'] = tips_df['tip']/tips_df['total_bill']
# 다변량 표현
# 성별에 따라 - 점의 색깔
# 값의 크기에 따라 - 점의 크기
fig = plt.figure()
axes = fig.add_subplot(1,1,1)

axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s= tips_df['tip_pct']*1000, 
    c= tips_df['sex'].map(colors),
    alpha=0.5
)
plt.show
```

```python
# 파이 차트 : 비율
# 성별 비율 비교

# 빈도수
sex_counts = tips_df['sex'].value_counts()

fig = plt.figure()
axes = fig.add_subplot(1,1,1) #

axes.pie(
    sex_counts,
    colors= ['blue','red'],
    labels =  sex_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.show
```

```python
fig = plt.figure()
colors = {"Female": "#ff00ad", "Male": "#000080"}
tips_df['tip_pct'] = tips_df['tip']/tips_df['total_bill']
# 다변량 표현
# 성별에 따라 - 점의 색깔
# 값의 크기에 따라 - 점의 크기
axes = fig.add_subplot(1,2,1)

axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s= tips_df['tip_pct']*1000, 
    c= tips_df['sex'].map(colors),
    alpha=0.5
)

# 빈도수
sex_counts = tips_df['sex'].value_counts()

axes2 = fig.add_subplot(1,2,2) #

axes2.pie(
    sex_counts,
    colors= ['blue','red'],
    labels =  sex_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.show
```
